# Question :-

## For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically.

## First import all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.metrics import accuracy_score
```

```python
# Reading the csv file

df=pd.read_csv("Iris.csv")
```

```python
df
```

```python
# Initializing the data on which the model is to be trained

x=df.iloc[:,1:5]
y=df.iloc[:,-1]
```

```python
# splitting the data into Train and Test data

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

## Initializing the Model

```python
model=dtc()
md=model.fit(xtrain,ytrain)
```

## Predicting the Values based on both xtest and ytest

```python
ypred=model.predict(xtest)
ypred_train=model.predict(xtrain)
```

## To show the Accuracy of the model

```python
accuracy_score(ypred,ytest)*100
```

```python
accuracy_score(ytrain,ypred_train)*100
```

# Visualizing the Decision Tree

```python
from sklearn import tree
```

```python
plt.figure(figsize=(22,20))
tree.plot_tree(model,feature_names=df.keys()[:-1],class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
               filled=True,rounded=True,fontsize=15)
plt.show()
```
